```python
import pandas as pd
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import association_rules

data = pd.read_csv('Day1.csv')
print(data)

encoded_data = pd.get_dummies(data)
print("Encoded Data= \n",encoded_data)

frequent_itemsets = fpgrowth(encoded_data, min_support=0.01, use_colnames=True)
print("\n Frequent Itemsets= \n",frequent_itemsets)

rules = association_rules(frequent_itemsets, metric="lift", min_threshold=1)
print("\n Frequent Itemsets= \n",rules.sort_values('lift', ascending=False).head(10))
```
